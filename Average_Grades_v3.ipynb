```python
# In this program we will learn about
# Modifying lists
# Deleting elements in list (POP)
# Clearing full list
# Removing specific elements in list


import math

#Create an empty list
grades = []


while True:
    print("\nEnter your choice")
    print("\n1. Enter Grades")
    print("2. Delete Grades")
    print("3. Update Grades")
    print("4. Clear Grades")
    print("5. Calculate Average")
    print("6. Check list of grades")
    print("7. Exit")

    choice = input()
    
    if not (len(choice) == 1 and choice.isnumeric()):
        print("\nPlease enter a valid choice\n")
        continue
    if int(choice) == 7:
        break
    if int(choice) == 1:
        while True:
            grade = input ("\nEnter Grades : ( e to \"exit\" )")
            if grade == "e":
                break
            else:
                grade = float(grade)
                grades.append(grade)
    if int(choice) == 6:
            print(grades)
    if int(choice) == 2:
        print("Enter the index for deletion of the element : ")
        d = int(input())
        grades.pop(d)
    if int(choice) == 3:
        print("Enter the index of number to be updated")
        u = int(input())
        print("Enter the value to be updated")
        up = float(input())
        grades[u] = up
    if int(choice) == 4:
        grades.clear()
    if int(choice) == 5:
        sum = 0
        for grade in grades:
            sum = sum + grade
        average = sum / len(grades)
        print(average)
```
